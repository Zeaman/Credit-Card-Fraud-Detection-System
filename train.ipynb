```python
# import installed packages
import joblib
import numpy as np
import pandas as pd
import opendatasets as od
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
# Specify the Kaggle dataset URL
kaggle_dataset_url = "https://www.kaggle.com/datasets/mlg-ulb/creditcardfraud"
# Download the Kaggle dataset
od.download(kaggle_dataset_url)
```

# START HERE

```python
# Upload the dataset
creditCard_dataset = ("creditcardfraud/creditcard.csv")
```

```python
# Upload the dataset to Pandas dataframe
creditCard_df = pd.read_csv(creditCard_dataset)
```

#### Interpreting:
#### -> Value "0" has a count of 284,315 occurrences.
#### -> Value "1" has a count of 492 occurrences
#### That indicated a highly biased dataset
#### Where:
#### 0 => Normal Transactions
#### 1 => Fraud Transactions

```python
# Assigning Classes in Pandas DataFrame
normal = creditCard_df[creditCard_df.Class == 0]
print(normal.shape)
fraud = creditCard_df[creditCard_df.Class == 1]
print(fraud.shape)
```

```python
# Print out the statistical measures of the above dataframe
normal.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
# Print out the values for both transactions after the dataset is cleaned and unbiased
creditCard_df.groupby('Class').mean()
```

```python
normal_sample = normal.sample(n = fraud.shape[0])
```

```python
creditCard_new_df = pd.concat([normal_sample, fraud], axis = 0)
```

```python
# Visualized the new datast/dataframe
creditCard_new_df.head()
```

```python
creditCard_new_df.tail()
```

```python
creditCard_new_df['Class'].value_counts()
```

```python
creditCard_new_df.groupby('Class').mean()
```

```python
# Splitting the new dataset/dataframe into input feature and target feature
x = creditCard_new_df.drop(columns = 'Class', axis = 1)
y = creditCard_new_df['Class']
```

```python
print(x)
```

```python
print(y)
```

```python
# Splitting the new dataset/df into training and testing data
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, stratify = y, random_state = 2)
```

```python
print(x.shape, x_train.shape, x_test.shape)
```

## Train the LR model using the above df

```python
# Define the LR model
model = LogisticRegression()
```

```python
# Train the LR model
model.fit(x_train, y_train)
```

## Evaluate the model

```python
# Accuracy score for the training data
x_train_prediction = model.predict(x_train)
training_data_accuracy = accuracy_score(x_train_prediction, y_train)
print("Accuracy of the training data: ", training_data_accuracy)
```

```python
# Accuracy score for test data
x_test_prediction = model.predict(x_test)
test_data_accuracy = accuracy_score(x_test_prediction, y_test)
print("Accuracy of the test data:", test_data_accuracy)
```

```python
# 
joblib.dump(model, 'best_model.joblib')
```
